### 2 Ejercicio

El fichero [quakes.csv](./datos/quakes.csv) recoge los datos de observaciones de movimientos sísmicos de magnitud superior a 4 en la escala de Richter  en un sector de la ciudad de Fiji desde 1964. 

1. Mostrar todos los datos con un único decimal. Para ello aplica la función __applymap__ .

```python
tabla_sismicos = pd.read_csv('./datos/quakes.csv', 
                             names = ['Latitud', 'Longitud', 'Profundidad','Magnitud', 'Estaciones'],
                             sep= ","
                            )
tabla_sismicos.head()
```

```python
#Vamos a sustituir todos los '-' del DataFrame por NaN ya que con la funcion applymap no nos va a dejar cambiar a un decimal si hay elementos string
tabla_sismicos1 = pd.read_csv('./datos/quakes.csv', 
                             names = ['Latitud', 'Longitud', 'Profundidad','Magnitud', 'Estaciones'],
                             sep= ",",
                             na_values= '-'
                            )
tabla_sismicos1.head()
```

```python
#Establecemos la funcion applymap con un lambda para sustituir por 1 decimal
tabla_sismicos = tabla_sismicos.applymap(lambda x: '%.1f'% x)
tabla_sismicos.head()
```

```python
#Otra forma de hacerlo es con replace, vamos a pedir ayuda para ver como utilizarla:
help(pd.DataFrame.replace)
```

```python
p = pd.DataFrame.replace(tabla_sismicos,
                         to_replace='-',
                         value=0)
p.head()
```

```python
# Sol:
tabla_sismicos1 = p.applymap(lambda x: '%.1f' % float(x))
tabla_sismicos1.head()
```

```python
#Por último, podemos hacerlo tambien mediante un bucle inicialmente:
def fun(x):
    if x == '-':
        return 0
    else:
        return float(x)
t = tabla_sismicos.applymap(fun)
t.head()
```

```python
# Sol:
tabla_sismicos1 = t.applymap(lambda x: '%.1f' % x)
tabla_sismicos1.head()
```
